## Introduction
Beatboxing is the art of vocal percussion. Beatboxers use it as a compelling form of expression within hip-hop culture and beyond, where performers use their mouths and voices to create complex musical sounds. While often associated with street performances and musical improvisation, the underlying acoustics and mechanisms of beatboxing present a fascinating area of scientific study. This exploration combines elements of linguistics, physics, and human physiology, offering insights into how the human vocal apparatus can mimic an array of percussive and musical instruments. 


At its core, beatboxing is about producing a wide range of sounds using only the human body's natural abilities. This involves precise control over various parts of the vocal tract, including the lips, tongue, throat, and nasal passages. The sounds produced can be broadly categorized into three types: percussive, tonal, and frictional.
1. Percussive Sounds: These mimic traditional drum sounds, like kicks and snares. Beatboxers produce these sounds by creating short, sharp pulses of air pressure, often using techniques similar to those found in linguistic plosive consonants like 'p', 't', and 'k'.
2. Frictional Sounds: These include hi-hats and cymbal sounds, produced by forcing air through a narrow aperture somewhere in the vocal tract, often the teeth and lips, creating high-frequency noise.
3. Tonal Sounds: While less common, some beatboxers can produce pitched tones, sometimes using their vocal folds to create bass lines or melodies. This requires the ability to sing and to modulate pitch precisely while performing other percussive elements.


### Motivation and Objective
This paper aims to investigate the acoustics and phonetics of beatboxing, specifically exploring the intricate ways in which beatboxers manipulate their articulatory organs to produce a broad spectrum of percussive and melodic sounds. Central to this exploration are the questions: 1. **How do beatboxers manipulate their articulators to produce a wide range of percussive sounds?** 2. **Are there variations in sound production between different beatboxers?**

The focus of this research will be on three foundational sounds in beatboxing: the drum, the hi-hat, and the snare. These elements are essential for constructing the rhythmic framework that characterizes beatboxing performances. To enrich our understanding and analysis, I have collected  data from three skilled beatboxers, specifically recording their production of these three fundamental sounds.

Additionally, recognizing the potential parallels between conventional speech sounds and beatboxing elements, I also recorded these beatboxers producing specific consonant sounds that are acoustically similar to their beatboxing counterparts. This approach allows for a **direct comparison between the beatboxing drum and the bilabial plosive [p], the hi-hat and the alveolar plosive [t], and other relevant consonantal sounds**. By examining both the unique articulatory traits of beatboxing and their relationships to more traditional phonetic elements, this paper seeks to understnad the sophisticated vocal techniques that beatboxers use.

**Hypothesis: There will exist variations in the same beatboxing sounds when produced by different beatboxers.**

## 1. The Kick Drum
*Linguistic description: Voiceless glottalic egressive labial stop*

In traditional drum kits, the kick drum provides the fundamental, deep tone that anchors the rhythm and can be felt as much as heard due to its low frequency. In beatboxing, replicating this sound effectively involves technique manipulation to mimic that deep, resonant thud without the use of an actual drum. The most common technique to produce a beatboxing kick drum sound involves a voiced or unvoiced bilabial plosive, typically represented as a [b] or [p] sound. 

### Data Visualization

```python
# Load images
img1 = mpimg.imread('p.png')
img2 = mpimg.imread('drum_Jack.png')
img3 = mpimg.imread('drum_Anura.png')
img4 = mpimg.imread('drum_Archan.png')

# Create a figure to hold the plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))  # 2 rows, 2 columns

# Display first image
axs[0, 0].imshow(img1)
axs[0, 0].axis('off')  # turn off axes for the first plot
axs[0, 0].set_title('p')

# Display second image
axs[0, 1].imshow(img2)
axs[0, 1].axis('off')  # turn off axes for the second plot
axs[0, 1].set_title('Jack')

# Display third image
axs[1, 0].imshow(img3)
axs[1, 0].axis('off')  # turn off axes for the third plot
axs[1, 0].set_title('Anura')

# Display fourth image
axs[1, 1].imshow(img4)
axs[1, 1].axis('off')  # turn off axes for the fourth plot
axs[1, 1].set_title('Archan')

fig.suptitle('Drum Sounds Visualization', fontsize=16, fontweight='bold')

plt.tight_layout()  
plt.show()
```

```python
# Load audio files
paths = ['p.wav','drum_Jack.wav','drum_Anura.wav','drum_Archan.wav']
audio_data = [librosa.load(p, sr=44100) for p in paths]  # sr=44100 to normalize sample rate

stfts = []
for data in audio_data:
    n_fft = min(len(data[0]), 2048)  # Ensuring n_fft does not exceed the signal length
    stft = librosa.stft(data[0], n_fft=n_fft)
    stfts.append(stft)

# Compute decibel levels from the amplitude (STFT)
db_levels = [librosa.amplitude_to_db(abs(stft)) for stft in stfts]

# Plotting
plt.figure(figsize=(14, 6))
labels = ['p', 'Jack', 'Anura', 'Archan']

for db, label in zip(db_levels, labels):
    db_mean = np.mean(db, axis=0)  # Simplifying plot by averaging across frequency bins
    times = librosa.frames_to_time(np.arange(db_mean.shape[0]), sr=44100, n_fft=n_fft)
    plt.plot(times, db_mean, label=label)

plt.title('Decibel Level Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Decibel Level (dB)')
plt.legend()
plt.show()
```

### Analysis

From both the graph and the table, we can see a clear distinction between the consonant [p] and the beatboxing kick drum. There are also apparent variations within the same sound produced by different beatboxers. Despite having the same place and manner of articulation, all three participants show a significantly higher maximum and minimum dB level compared to [p]. This suggests that the beatboxing drum sounds are produced with greater force and air pressure, leading to louder sounds. The production of the kick drum sound involves the lips closing together and the velum rising to create a seal in the mouth. The vocal folds close together, then rise to create high pressure behind the lips. When the lips separate, the pressure is released with an explosive sound. Significantly more air is being built up in the mouth for a beatboxing drum sound. 

The average decibel level provides an overall sense of the loudness of the sound. All three beatboxing samples exhibit higher average dB levels compared to [p]. This indicates that beatboxers sustain a higher volume throughout the sound production, which is characteristic of a performance style aiming to mimic powerful drum sounds.

Another feature to focus on is the attack time: How quickly does the sound reach its maximum loudness? This can indicate how sharply a sound begins. For beatboxing sounds, a sharper attack might be indicative of techniques intended to mimic percussive instruments. It's surprising to see how that the attack time for all three beatboxers almost match up exactly. The attack time for the beatboxing drum sound is about 0.01s, and after that, it starts to decay. However, the sound [p] starts off with the maximum loudness then fades off. 

In music production, the precise alignment of sound attacks, particularly in percussion, is critical for the rhythmic quality of a piece. Beatboxers’ ability to synchronize their attack times across different sounds suggests that they can adapt their rhythmic output to closely resemble that of actual drum kits, while maintaining the tempo essential to effective musical performance.

The manner and place of articulation for this sound  closely resemble that of the English alveolar stop [t], commonly heard in English and many other languages.

Upon examining the waveform, it is immediately evident that the sound [t] is less intense and exhibits less acoustic energy compared to the beatboxing hi-hat sound. This increased intensity suggests that  beatboxers apply greater subglottal pressure during sound production, which contributes to a louder and more forceful sound output typical of general percussive instruments.

The analysis reveals that the attack time is almost identical for both [t] and Jack's beatboxing hi-hat. This similarity in attack times indicates a precise control over the articulation mechanism, allowing Jack to replicate the sharp, abrupt onset typical of a hi-hat cymbal. The decay patterns, where the sound reduces in amplitude post-attack, also align closely, demonstrating skillful modulation of the sound's termination phase to mimic the quick dampening of a hi-hat.

Two notable differences provide deeper insights into the distinct nature of these sounds:

- Sound Level Intensity: Jack's beatboxing hi-hat sound maintains a higher sound level than the linguistic [t] throughout its duration. This higher intensity level is indicative of sustained subglottal pressure, which is a technique used by beatboxers to amplify their sounds and achieve a more impactful auditory effect reminiscent of real percussion instruments.

- Sustain and Decay Duration: The sustain and decay phases for Jack’s beatboxing hi-hat are shorter compared to those of the linguistic [t]. Jack purposefully shortens the sustain and cuts off the decay phase quickly after the peak is reached. This technique results in a crisper and cleaner sound, effectively emulating the sharp, clipped quality of an actual hi-hat cymbal being struck and muted rapidly.

Unlike the drum sound, the hi-hat exhibits a variety of differences in the technique used among the three beatboxers. One notable difference is that the attack time and release time for Anura is significantly longer than the rest. This is because Anura's uses a different sound hi-hat sound than Jack and Archan. Her sound resembles the **Voiceless alveolar sibilant affricate: [ts]** It is produced by first stopping the air flow entirely, then directing it with the tongue to the sharp edge of the teeth, causing high-frequency turbulence. 

The production of [ts] involves a two-part articulatory process where the airflow is first completely stopped (the occlusive phase) and then released (the fricative phase) through a narrow channel formed by the tongue and the sharp edges of the teeth.

## 3. The Snare
*Linguistic description: Voiceless glottalic egressive labial affricate* / *Voiceless pulmonic ingressive lateral velar affricate*



Mimicking the sharp, crisp sound of a snare drum, beatboxers use various vocal techniques to achieve different snare sounds. There's more variability in the snare sound than in the previous two sounds discussed. In fact, the data collected from the three beatboxers represent three different varieties and techniques on the snare sound. This is because snares in beatboxing can be produced using a variety of vocal techniques involving different parts of the mouth and throat. For example, beatboxers use techniques such as the rimshot, inward or outward K-snare, PF-snare, and others. Each technique utilizes different tongue positions, lip actions, and breath control, contributing to the distinctive sound qualities of each snare type.

1. Anura used a *k snare*, which is a voiceless pulmonic ingressive lateral velar affricate. Anura's k snare is characterized by the ingressive sound created by drawing air inward rather than pushing it out. This inward technique is combined with a velar closure, where the back of the tongue meets the soft palate, followed by a lateral release that allows the air to escape around the sides of the tongue. This method produces a deep, resonant snare that has a slightly hollow quality compared to the more traditional pf snare. 

2. Jack used a *pf snare*, which is a voiceless glottalic ingressive labial affricate. The sound is produced by a forceful ingression of air through a narrow aperture formed between the lips and teeth, creating a high-pressure sound that closely mimics the sharp snap of a traditional snare drum. This technique involves a quick build-up and release of air pressure, which contributes to the crispness and sharpness of the sound. The "pf" sound is articulated with the lips starting closed, then abruptly opening with a burst of air that creates the snare-like crack.

3. Archan used his own snare sound, which is not a traditional / documented snare sound. His snare technique is similar to that of Anura's hi-hat, by forcing air through a [ts] mouth positioning. His tongue is positionted at the palate where a [t] sound would be.

On the other hand, Jack and Archan have longer attack times because their techniques involve more complex articulatory maneuvers that delay the full release of air, creating a build-up of pressure before the sound peaks. Jack uses a "pf" snare. This technique involves closing the lips tightly and then releasing them with force, creating a sharp, explosive sound. The "pf" sound inherently has a quick attack due to the nature of the articulation, but it can be extended slightly by delaying the release of the lips. Archan's approach, which is similar to producing an aspirated [ts] sound, involves more intricate tongue and palate interactions. His technique can be described as forcing air through an aspirated position where the tongue is close to the palate, ready to release a [t] sound, followed by a forceful aspiration. This process involves both a stop and a fricative element, which naturally extends the attack phase. The preparation and execution of this combined articulatory movement result in a more pronounced buildup and a dramatic release, contributing to a longer attack time.
 

The first formant (F1) provides an insight into the resonant frequency of the vocal tract configuration during sound production. Archan, with an F1 of 1691.04 Hz, shows a higher pitch or tonal quality compared to the other snares. This might be attributed to his unique technique of forcing air through an aspirated [ts], which possibly alters the vocal tract's shape more dramatically than the techniques used by Jack and Anura.

Jack and Anura show lower F1 frequencies than Archan’s but are significantly higher than the [k] sound, indicating that while Jack and Anura's snares are tonally higher than a standard [k] sound, they are not as high-pitched as Archan's. This could reflect their respective techniques’ influence on vocal tract configuration—Jack's "pf" sound requiring a tight labial closure and sudden release, and Anura’s "k" snare utilizing a velar closure with lateral release.
